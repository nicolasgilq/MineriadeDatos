# Exercise 18.1

Analyze the writing patterns of each student

```python
from sklearn.feature_extraction.text import CountVectorizer

# define X and y
data['Sex'] = data['Sexo'].map({'H': 1, 'M': 0})
data["T1"].fillna("", inplace=True)
data["T2"].fillna("", inplace=True)
data["T3"].fillna("", inplace=True)
data["T4"].fillna("", inplace=True)
data["T5"].fillna("", inplace=True)
data["T6"].fillna("", inplace=True)
feature_cols = ['T1', 'T2', 'T3', 'T4', 'T5', 'T6']
X = data['T1'] + data['T2'] + data['T3'] + data['T4'] + data['T5'] + data['T6']
y = data['Sex']

# Análisis de palabras
vect1 = CountVectorizer(lowercase=False)
X_dtm1 = vect1.fit_transform(X)
temp = X_dtm1.todense()
print(vect1.vocabulary_)

```

```python
# last 50 features
print(vect1.get_feature_names()[-150:-100])

```

```python
# Palabra más usada
print(vect1.get_feature_names()[X_dtm1.todense()[0].argmax()])

```

```python
# Análisis de frases
vect2 = CountVectorizer(ngram_range=(2, 4))
X_dtm2 = vect2.fit_transform(X)
print(vect2.vocabulary_)

```

```python
# last 50 features
print(vect2.get_feature_names()[-1050:-950])

```

```python
# Frase más usada
print(vect2.get_feature_names()[X_dtm2.todense()[0].argmax()])
```

# Exercise 18.2

Evaluate the similarities of the homeworks of the students

tip: https://github.com/orsinium/textdistance

```python
import textdistance

# X1 = data['T1']  # Similitudes en la tarea 1
X1 = data['T1'] + data['T2'] + data['T3'] + data['T4'] + data['T5'] + data['T6']  # Similitudes en todas las tareas

matriz = np.zeros((len(X1), len(X1)))
for ia in range(0, len(X1)):
    for ja in range(0, len(X1)):
        if ia == ja:
            matriz[ia][ja] = 0
        else:
            matriz[ia][ja] = textdistance.hamming.normalized_similarity(X1[ia], X1[ja])

print("% de similitud: ", matriz.max()*100)
print("Ubicación: ", np.where(matriz == matriz.max()))
```
